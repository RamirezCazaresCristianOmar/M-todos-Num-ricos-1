```python
# Importar las bibliotecas necesarias
from sympy import Symbol, expand
from sympy.plotting import plot
import matplotlib.pyplot as plt
```

```python
def lagrange_interpolation(xi, fi):
    """
    Calcula el polinomio de interpolación de Lagrange dados los puntos (xi, fi).

    Argumentos:
    xi -- Lista de valores x (abscisas)
    fi -- Lista de valores f(x) (ordenadas)

    Retorna:
    - El polinomio de interpolación de Lagrange como expresión simbólica
    """
    # Verificar que las listas tengan la misma longitud
    if len(xi) != len(fi):
        raise ValueError("Las listas de valores x y f(x) deben tener la misma longitud")

    # Crear variable simbólica
    x = Symbol('x')

    # Inicializar el polinomio
    polinomio = 0
    n = len(xi)

    # Calcular el polinomio de Lagrange
    for i in range(n):
        # Calcular el término Li(x)
        Li = 1
        for j in range(n):
            if i != j:
                Li *= (x - xi[j])/(xi[i] - xi[j])
        # Multiplicar por fi y sumar al polinomio
        polinomio += fi[i]*Li

    # Expandir el polinomio para simplificarlo
    return expand(polinomio)

def plot_lagrange(xi, fi, polinomio, titulo="Interpolación de Lagrange"):
    """
    Grafica los puntos originales y el polinomio de interpolación.

    Argumentos:
    xi -- Lista de valores x
    fi -- Lista de valores f(x)
    polinomio -- Polinomio de interpolación (expresión simbólica)
    titulo -- Título del gráfico (opcional)
    """
    # Crear la figura
    plt.figure(figsize=(10, 6))

    # Graficar los puntos originales
    plt.plot(xi, fi, 'ro', label='Puntos originales', markersize=10)

    # Graficar el polinomio
    p = plot(polinomio, show=False)
    plt.plot(*p[0].get_points(), 'b-', label='Polinomio de Lagrange')

    # Configurar el gráfico
    plt.grid(True)
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

# Ejemplo de uso
def ejemplo_interpolacion():
    """
    Ejemplo de uso de las funciones de interpolación de Lagrange.
    """
    # Datos de ejemplo
    xi = [1.0, 1.3, 1.6, 1.9, 2.2]
    fi = [0.7651977, 0.6200860, 0.4554022, 0.2818186, 0.1103623]

    # Calcular el polinomio de interpolación
    polinomio = lagrange_interpolation(xi, fi)

    # Mostrar el polinomio
    print("Polinomio de interpolación de Lagrange:")
    print(polinomio)

    # Graficar el resultado
    plot_lagrange(xi, fi, polinomio)

if __name__ == "__main__":
    ejemplo_interpolacion()
```
