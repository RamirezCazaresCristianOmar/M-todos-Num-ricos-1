```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#definimos la funcion para calcular el polinomio de Newton
def newton_p(x, x_datos, y_datos):

#Calcula el valor del polinomio de Newton para un valor dado de x.
#argumentos:
#x:El valor de x para el cual se quiere evaluar el polinomio.
#x_datos:Un array de valores de x de la tabla de datos.
#y_datos:Un array de valores de y de la tabla de datos.

 n = len(x_datos)
 y = np.zeros(n)
 y[0] = y_datos[0]

#calculamos la diferencias divididas
 for i in range(1,n):
   for j in range(n-i,0,-1):
    y_datos[j] = (y_datos[j]-y_datos[j - 1]) / (x_datos[j] - x_datos[j - i])

#calculamos el valor del polinomio de Newton
 for i in range(1,n):
  y[i] = y[i - 1] + y_datos[i] * np.prod(x - x_datos[:i])

  return y[-1]

#Definimos los datos de la tabla

 x_datos = np.array([0, 1, 2, 3, 4])

 y_datos = np.array([1, 2, 5, 10, 17])

#calcula los valores del polinomio de Newton para los valores de x
 y = np.array([newton_p(xi,x_datos,y_datos)for xi in x])
```

```python
#Generamos valores de x para la grafica
x = np.linspace(x_datos[0],x_datos[-1],100)
```

```python
#Grafica los datos de la tabla y el polinomio de Newton
plt.plot(x_datos,y_datos,'o',label='Datos')
plt.plot(x, y,'-',label='Polinomio de Newton')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("Interpolacion de newton")
plt.show()
```
